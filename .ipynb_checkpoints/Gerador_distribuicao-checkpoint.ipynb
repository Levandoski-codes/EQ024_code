```python
import random
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import math
import pandas as pd
```

# Curva
### Gerador de distribuição aleatória, baseada na curva normal

```python
numero_peneiras = 7 #Escolhendo o número de peneiras a ser utilizadas
razao = (2)**(1/4) #Escolhendo a razão entre os diâmetros das peneiras Raiz quarta
abertura_1_penei = 8000 #Abertura da primeira peneira em microns 
media_dia = 4000 #Usando a media da distribuição normal,
                 #baseada no dado fornecido pela literatura
massa_total = 5680 #gramas 
var = 0.3 #Uma variação da curva normal para gerar os dados
```

```python
2*(2)**(1/4)
```

```python
random.seed(10)
dist_norm = []
for _ in range(massa_total):
    dist_norm.append(random.normalvariate(media_dia,
                                          media_dia*var))
```

## Classificação do diâmetro seguindo a razão inserida

```python
anterior = float(abertura_1_penei)
for _ in range(numero_peneiras):
        agora = anterior/razao
        print(f'-{anterior:9.7}\t+{agora:9.7}')
        anterior = agora
```

# Tamanho das peneiras escolhidas em Microns

```python
peneiras = [
    6680, #Mesh 3
    4699, #Mesh 4
    3327, #Mesh 6
    2362, #Mesh 8
    1651, #Mesh 10
    1168, #Mesh 14
    833,  #Mesh 20
    0     #Bandeja 
]
peneiras = np.asarray(peneiras, dtype='int')
```

```python
penei = {} #Criando um dicionário com os dados 
bins = [] #Será usado como base para distribuição do histograma
for j in range(len(peneiras)-1):
    lista =[]
    for i in dist_norm:
        if peneiras[j] > i >=peneiras[j+1]:
            lista.append(i)
    bins.append(peneiras[j])
        
    penei.update({str(peneiras[j]): lista})
    
for i in dist_norm:
        if i < peneiras[-2]:
            lista.append(i)
bins.append(peneiras[-1])
penei.update({str(peneiras[-1]): lista})
```

```python
#Calculando as médias de cada peneira
x = []
y = []
for nome in penei:
    x.append(int(nome))
    y.append(len(penei[nome]))
    
x = np.asarray(x[::-1]) #Isso é um array com a média de diâmetro
                        #presente em cada uma das peneiras
y = np.asarray(y[::-1]) #Massa de material presente em cada peneira
frac = (np.cumsum(y)/ np.cumsum(y)[-1])*100 #Calculando a fração acumulada
bins = bins[::-1]
```

```python
df = pd.DataFrame((peneiras, 100*y/massa_total, frac[::]), dtype='int')
df.rename(columns={0:'-$\infty$\t+3', 1:'-3\t+4',
                   2:'-4\t+6', 3:'-6\t+8',
                   4:'-8\t+10', 5:'-10\t+14',
                   6:'-14\t+20',7:'-20\t+$\infty$'}, inplace=True)
df.rename(index={0:'Diâmetro peneira ($\mu$m)', 1:'Frequência Ponderal', 2:'Fração Acumulada'}, inplace=True)
```

### Plot da distribuição

```python
fig = plt.figure(figsize=(15,8), dpi=200)
plt.plot(peneiras[::-1], frac, marker='.', color='red')
plt.xticks(bins)
plt.title('Distribuição granulométrica')
plt.xlabel('Tamanho da peneira ($\mu$m)')
plt.ylabel('Fração acumulada')
plt.show()
```

```python
fig = plt.figure(figsize=(15,8), dpi=200)
plt.xlim((math.ceil(0), math.floor(abertura_1_penei*razao)))
plt.xticks(bins)
plt.hist(dist_norm, bins=bins)
plt.title('Histograma')
plt.xlabel('Tamanho da peneira ($\mu$m)')
plt.ylabel('Massa de material (g)')
plt.show()
```


### Plot ddo histograma da distribuição, baseado no número de peneiras escolhidas

```python
fig = plt.figure(figsize=(15,8), dpi=200)
ax = plt.axes()
plt.ylabel("Fração normalizada")
values, base, _ = plt.hist(dist_norm, bins = bins, density=True,
                           alpha = 0.5, color = "green",
                           range = range, label = "Histograma")
ax_bis = ax.twinx()
values = np.append(values,0)
ax_bis.plot(base, np.cumsum(values)/ np.cumsum(values)[-1]*100,
            color='darkorange', marker='*', linestyle='-',
            markersize = 1, label = "Histograma Cumulativo" )
plt.title('Distribuição granulométrica vs Tamanho da peneira ($\mu$m)')
plt.xlabel('Tamanho da peneira ($\mu$m)')
plt.ylabel('Fração acumulada')
plt.xticks(bins)
ax_bis.legend();
ax.legend();
plt.show()
```

# Modelos

```python
#Calculando as médias de cada peneira
di = []
fraction = []
for nome in penei:
    di.append(np.median(penei[nome]))
    fraction.append(len(penei[nome]))
    
di = np.asarray(di[::-1]) #Isso é um array com a média de diâmetro
                        #presente em cada uma das peneiras
fraction = np.asarray(y[::-1]) #Massa de material presente em cada peneira
```

```python
X = frac/100
di = di
```

```python
%matplotlib inline
```

```python
# %matplotlib qt
```

```python

global d_linha
d_linha = 3737

def func(x, b):
    return 1 - np.exp(-(x/d_linha)**b)

xdata = di
ydata = X

fig = plt.figure(figsize=(15,8), dpi=200)

plt.plot(xdata, ydata, '--', label='Experimental', color='blue', marker='o')
popt, pcov = curve_fit(func, xdata, ydata)

# axis = np.linspace(min(xdata), max(xdata), 100)
axis = xdata

ax = plt.plot(axis, func(axis, *popt), 'r-',
         label=f"X = 1 - exp[-(d$_i$/d)$^n$]\nn: {popt[0]:5.4}\nd': {d_linha}\n$\sigma$: {np.sqrt(np.diag(pcov))[0]:5.4}",
         color='red', marker='.')

plt.title('Modelo de distribuição Rosin-Ramler-Bennet')
plt.xlabel('di ($\mu$m)')
plt.axhline(0.632, color='black')
plt.axvline(d_linha,color='black')
plt.ylabel('X')
plt.legend()
plt.show()
```

```python

global K
K = 5288

def func_g(x, b):
    return (x/K)**b

xdata = di
ydata = X

fig = plt.figure(figsize=(15,8), dpi=200)

plt.plot(xdata, ydata, '--', label='Experimental', color='blue', marker='o')
popt_g, pcov_g = curve_fit(func_g, xdata, ydata)

# axis = np.linspace(min(xdata), max(xdata), 100)
axis = xdata

ax = plt.plot(axis, func_g(axis, *popt_g), 'r-',
         label=f"X = (d$_i$/K)$^m$\nm: {popt_g[0]:5.4}\nK: {K}\n$\sigma$: {np.sqrt(np.diag(pcov_g))[0]:5.4}",
         color='red', marker='.')

plt.title('Modelo de distribuição Gates-Gaudin-Schumann')
plt.xlabel('di ($\mu$m)')
plt.axhline(1, color='black')
plt.axvline(K, color='black')
plt.ylabel('X')
plt.legend()
plt.show()
```
